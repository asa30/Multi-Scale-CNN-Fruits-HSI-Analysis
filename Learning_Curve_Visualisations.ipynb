```python
import re 
import matplotlib.pyplot as plt
```

```python
pattern1 = r"loss: ([0-9.]+(?:e[+-]?[0-9]+)?) - accuracy: ([0-9.]+) - val_loss: ([0-9.]+(?:e[+-]?[0-9]+)?) - val_accuracy: ([0-9.]+)"
```

```python
# 8s 877ms/step - loss: 0.9973 - accuracy: 0.5556
pattern2 = r"[0-9.]+s [0-9.]+[m]?s/step - loss: ([0-9.]+(?:e[+-]?[0-9]+)?) - accuracy: ([0-9.]+)"
```

```python
def visualise_training_output(output, title, pattern):
    matches = re.findall(pattern, output)
    losses = [float(match[0]) for match in matches]
    accuracies = [float(match[1]) for match in matches]
    losses = losses[:100]
    accuracies = accuracies[:100]
    # plot them
    fig, ax = plt.subplots(2)
    fig.suptitle(title)
    plt.subplots_adjust(hspace=0.5)  # Add space between subplots
    ax[0].plot(losses)
    ax[0].set_title("Training Loss")
    ax[0].set_xlabel("Epoch")
    ax[0].set_ylabel("Loss")
    ax[1].plot(accuracies)
    ax[1].set_title("Training Accuracy")
    ax[1].set_xlabel("Epoch")
    ax[1].set_ylabel("Accuracy")
```

```python
def visualise_training_output_comparison(output1, output2, l1, l2, title, pattern):
    matches1 = re.findall(pattern, output1)
    losses1 = [float(match[0]) for match in matches1]
    accuracies1 = [float(match[1]) for match in matches1]
    losses1 = losses1[:100]
    accuracies1 = accuracies1[:100]
    matches2 = re.findall(pattern, output2)
    losses2 = [float(match[0]) for match in matches2]
    accuracies2 = [float(match[1]) for match in matches2]
    losses2 = losses2[:100]
    accuracies2 = accuracies2[:100]
    # plot them
    fig, ax = plt.subplots(2)
    fig.suptitle(title)
    plt.subplots_adjust(hspace=0.5)  # Add space between subplots
    ax[0].plot(losses1, label=l1)
    ax[0].plot(losses2, label=l2)
    ax[0].set_title("Training Loss")
    ax[0].set_xlabel("Epoch")
    ax[0].set_ylabel("Loss")
    ax[0].legend()
    ax[1].plot(accuracies1, label=1)
    ax[1].plot(accuracies2, label=2)
    ax[1].set_title("Training Accuracy")
    ax[1].set_xlabel("Epoch")
    ax[1].set_ylabel("Accuracy")
```

## HS_CNN

```python
avocado_nir_hscnn_output = open("TRAIN/job-58625.output", "r").read()
avocado_vis_hscnn_output = open("TRAIN/job-58626.output", "r").read()
kiwi_nir_hscnn_output = open("TRAIN/job-58623.output", "r").read()
kiwi_vis_hscnn_output = open("TRAIN/job-58624.output", "r").read()
```

```python
visualise_training_output(avocado_nir_hscnn_output, "Avocado NIR HSCNN", pattern2)
visualise_training_output(avocado_vis_hscnn_output, "Avocado VIS HSCNN", pattern2)
visualise_training_output(kiwi_nir_hscnn_output, "Kiwi NIR HSCNN", pattern2)
visualise_training_output(kiwi_vis_hscnn_output, "Kiwi VIS HSCNN", pattern2)
```
